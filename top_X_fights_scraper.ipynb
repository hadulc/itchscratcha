```python
import requests
import re

url = "https://www.fightmatrix.com/past-events-search/?txt=ufc"

response = requests.get(url)


```

```python

event_paths = re.findall(r"event/.*?/\'", response.text)
unique_paths = list(set(event_paths))
unique_paths[:20]  
```

```python
from datetime import datetime

month_to_number = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July': 7,
    'August': 8,
    'September': 9,
    'October': 10,
    'November': 11,
    'December': 12
}



def extract_from_event(event_path):
    response = requests.get(f"https://www.fightmatrix.com/{event_path}")
    ranks = re.findall(r'\[\#.*?\]', response.text)
    ranks = list(set(ranks))
    def clean_rank(rank):
        rank = rank[2:-1].split(" ")
        rank = int(rank[0]), rank[1]
        
    ranks = [clean_rank(rank) for rank in ranks]

    date = re.findall(r'day, .*?<', response.text)[0]
    date = date.split(", ")[1].split(" ")
    month, day, year = date[0], int(date[1][:-2]), int(date[2][:-1])
    month = month_to_number[month]
    event_date = datetime(year, month, day)
    return event_date, ranks
```

```python
for event_path in unique_paths[:5]:
    print(extract_from_event(event_path))
```
