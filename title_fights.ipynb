```python
import pandas as pd

df = pd.read_csv("data/title_fights.csv")

df
```

```python
pivot = df.pivot_table(index='year', columns='weight', values='winner', aggfunc='count')
pivot.index.name = 'year'

for weight in pivot.columns:
    first_year = min(year for year in pivot.index if not pd.isna(pivot.loc[year, weight]))
    pivot.loc[pivot.index > first_year, weight] = pivot.loc[pivot.index > first_year, weight].fillna(0)
    
pivot["sum"] = pivot.sum(axis=1)
pivot["avg"] = round(pivot.drop("sum", axis=1).mean(axis=1),1)
pivot.to_csv('data/title_fights_per_year.csv')
pivot
```
